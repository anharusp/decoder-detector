```python
train_dict = []
test_dict = []
```

```python
def dict_changer(current_dict):
    if 'uri' in current_dict.keys():
        for k in current_dict['uri'].keys():
            current_dict['uri_'+k] = current_dict['uri'][k]
    
    if 'headers' in current_dict.keys():
        for header in current_dict['headers']:
            current_dict[header[0]] = header[1]
            
    del current_dict['uri']
    del current_dict['headers']
    return current_dict
```

```python
for path, folder, files in os.walk('.'):
    if ('parsed' in path and 'gwt' not in path and 'perekrestok_base64' not in path):
        for (i, fname) in enumerate(files):
            print(path+'/'+fname)
            with open(path+'/'+fname) as request_file:
                request = request_file.read()
            
            request = json.loads(request)
            request['name'] = path+'/'+fname
            if 'train' in path:
                train_dict.append(dict_changer(request))
            else:
                test_dict.append(dict_changer(request))
```
